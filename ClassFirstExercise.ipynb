```python
import sys

acc_user = "Riyaan"
balance = 1000

print("Welcome Mr.", acc_user)

running = True

while running:
  print("\n 1.Deposit")
  print("\n 2.Withdraw")
  print("\n 3.CheckBalance")
  print("\n 4.Exit")

  choice = input("Enter your choice: ")

  if choice == "1":
      amount = float(input("Deposit amount:"))
      balance+=amount
      checkBal = input("Do you want to chek balance(Y/N):")
      if checkBal == "Y":
          print("Balance:",balance)

  elif choice == "2":
      amount = float(input("Withdraw amount:"))
      if amount <= balance:
          print("Amount Successfully Withdrawed")
          balance-=amount
          checkBal = input("Do you want to chek balance(Y/N):")
          if checkBal == "Y":
               print("Balance:",balance)
      else:
          print("Insufficient Balance")
  elif choice == "3":
      print("Your Balance amount is",balance)
  elif choice == "4":
      print("Thank You Mr.", acc_user)
      running = False 
  else:
      print("Invalid choice. Try again.")
```
