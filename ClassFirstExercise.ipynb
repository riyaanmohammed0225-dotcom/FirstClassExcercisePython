```python
import sys

acc_user = "Riyaan"
balance = 1000

print("Welcome Mr.", acc_user)

running = True

while running:
  print("\n 1.Deposit")
  print("\n 2.Withdraw")
  print("\n 3.CheckBalance")
  print("\n 4.Exit")

  choice = input("Enter your choice: ")

  if choice == "1":
      amount = float(input("Deposit amount:"))
      balance+=amount
      checkBal = input("Do you want to chek balance(Y/N):")
      if checkBal == "Y":
          print("Balance:",balance)

  elif choice == "2":
      amount = float(input("Withdraw amount:"))
      if amount <= balance:
          print("Amount Successfully Withdrawed")
          balance-=amount
          checkBal = input("Do you want to chek balance(Y/N):")
          if checkBal == "Y":
               print("Balance:",balance)
      else:
          print("Insufficient Balance")
  elif choice == "3":
      print("Your Balance amount is",balance)
  elif choice == "4":
      print("Thank You Mr.", acc_user)
      running = False
  else:
      print("Invalid choice. Try again.")
```

The above program is for Simple Banking where the account user is predefined.



"acc_user"="Riyaan" "Balance"="1000"


We use the running = true to keep the program running until the user exits the program.



1.I have created 4 choices where the user can Deposit,Withdraw,Check Balance or Exit

2.when he chooses the 1 or 2 or 3 he can deposit or withdraw or checkbalance

3.Until the chooses the 4th choice the looping does not stop

4.After user choose the loop ends and exits the program
