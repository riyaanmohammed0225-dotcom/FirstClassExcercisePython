```python
print("Welcome")

accounts = {
    "Riyaan": ["1234", 1000],
    "Ahmed": ["1111", 2000],
    "Ali": ["2222", 1500]
}
def create_account():
    name = input("Enter your Name: ")
    pin = input("Create PIN: ")
    balance = float(input("Enter Initial Balance: "))

    accounts[name] = [pin, balance]
    print("Account created successfully!\n")
def login_process():
    name = input("Enter Name: ")
    pin = input("Enter PIN: ")

    if name in accounts and accounts[name][0] == pin:
        print("Successfully Logged In")
        bank_menu(name)
    else:
        print("No Account Found or Wrong PIN")
def bank_menu(name):
    running = True
    while running:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            deposit(name)
        elif choice == "2":
            withdraw(name)
        elif choice == "3":
            check_balance(name)
        elif choice == "4":
            print("Thank You", name)
            running = False
        else:print("Invalid choice. Try again.")

def deposit(name):
    amount = float(input("Deposit amount: "))
    accounts[name][1] += amount
    print("Amount Deposited Successfully")

def withdraw(name):
    amount = float(input("Withdraw amount: "))
    if amount <= accounts[name][1]:
        accounts[name][1] -= amount
        print("Amount Successfully Withdrawn")
    else:
        print("Insufficient Balance")
def check_balance(name):
    print("Your Balance is:", accounts[name][1])
def main():
    login = input("Do you have an account? (Yes/No): ")
    if login.lower() == "no":
        create_account()
    login_process()

main()
```

## Small Explanation of the Program

This program is a **simple banking system** written in Python using functions.

* It first prints **Welcome**.
* It uses a **dictionary (`accounts`)** to store user name, PIN, and balance.
* It asks whether the user has an account.
* If not, it creates a new account.
* Then it checks the **login details (name and PIN)**.
* If correct, it shows a menu:

  * Deposit
  * Withdraw
  * Check Balance
  * Exit
* The menu repeats until the user exits.

Functions are used to make the program **clean and organized** instead of writing everything in one block.
