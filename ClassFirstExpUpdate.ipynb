```python
print("Welcome")
acc_user = {
    "Riyaan": ["1234", 1000],
    "Ahmed": ["1111", 2000],
    "Ali": ["2222", 1500]
}

login = input("Do you have an account? (Y/N): ")
if login.lower() == "n":
    name = input("Enter your Name: ")
    pin = input("Create PIN: ")
    balance = float(input("Enter Initial Balance: "))
    acc_user[name] = [pin, balance]
    print("Account created successfully!\n")

name = input("Enter Name: ")
pin = input("Enter PIN: ")

if name in acc_user and acc_user[name][0] == pin:
    print("Successfully Logged In")

    running = True
    while running:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            amount = float(input("Deposit amount: "))
            acc_user[name][1] += amount
            print("Amount Deposited Successfully")

        elif choice == "2":
            amount = float(input("Withdraw amount: "))
            if amount <= acc_user[name][1]:
                acc_user[name][1] -= amount
                print("Amount Successfully Withdrawn")
            else:
                print("Insufficient Balance")

        elif choice == "3":
            print("Your Balance is:", acc_user[name][1])

        elif choice == "4":
            print("Thank You", name)
            running = False

        else:
            print("Invalid choice. Try again.")

else:
    print("No Account Found or Wrong PIN")
```

The above program is for Simple Banking where the account user is predefined.


We use the running = true to keep the program running until the user exits the program.


1.I have created 4 choices where the user can Deposit,Withdraw,Check Balance or Exit

2.when he chooses the 1 or 2 or 3 he can deposit or withdraw or checkbalance

3.Until the chooses the 4th choice the looping does not stop

4.After user choose the loop ends and exits the program

THIS PROGRAM CONSIST OF VARIABLES,INPUT STATEMENT,OUTPUT STATEMENT AND LOOPING STATEMENT.
