## SIMPLE BANKING SYSTEM USING THE PYTHON *('if-else statements')*

* This Program has Predefined account and pin, but we can also Create New account.

* A Variable name *("acc_user")* is declared where all the accounts are saved.

* A Login input is used to Ask *('Do you have account?(Y/N)')*.Based on the answer the next process will be started.

* If the Answer is *'N'*, we can create a new account. The account will be saved in the variable *"acc_user"*.After Saving We can enter that name to access the other features.

* When the user enters the *"Name"* and *"Pin"*, if the user account is present in the Variable *"acc_user"*, it asks 4 choices, else It Shows *("No Account Found or Wrong PIN")*

* The 4 choices for Banking are

    * Deposit - To pay some amount (amount addition)
    * Withdraw - To take some amount (amount subtraction)
    * Check Balance - To check the balance amount
    * Exit

* Other than these 4 choices it shows a warning message - *"Invalid choice"*.

NOTE - *THIS PROGRAM CONSIST OF VARIABLES,INPUT STATEMENT,OUTPUT STATEMENT,CONDITIONAL STATEMENT AND LOOPING STATEMENT*.

```python
print("Welcome")
acc_user = {
    "Riyaan": ["1234", 1000],
    "Ahmed": ["1111", 2000],
    "Ali": ["2222", 1500]
}

login = input("Do you have an account? (Y/N): ")
if login.lower() == "n":
    name = input("Enter your Name: ")
    pin = input("Create PIN: ")
    balance = float(input("Enter Initial Balance: "))
    acc_user[name] = [pin, balance]
    print("Account created successfully!\n")

name = input("Enter Name: ")
pin = input("Enter PIN: ")

if name in acc_user and acc_user[name][0] == pin:
    print("Successfully Logged In")

    running = True
    while running:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            amount = float(input("Deposit amount: "))
            acc_user[name][1] += amount
            print("Amount Deposited Successfully")

        elif choice == "2":
            amount = float(input("Withdraw amount: "))
            if amount <= acc_user[name][1]:
                acc_user[name][1] -= amount
                print("Amount Successfully Withdrawn")
            else:
                print("Insufficient Balance")

        elif choice == "3":
            print("Your Balance is:", acc_user[name][1])

        elif choice == "4":
            print("Thank You", name)
            running = False

        else:
            print("Invalid choice. Try again.")

else:
    print("No Account Found or Wrong PIN")
```
